```python
# Sample correlation matrix (use your dataset's actual correlation matrix)
correlat = data[['carat', 'depth', 'table', 'price']].corr()

# Flatten the correlation matrix and filter for unique pairs only
correlation_pairs = []
for i in range(len(correlat.columns)):
    for j in range(i + 1, len(correlat.columns)):
        correlation_pairs.append((correlat.columns[i], correlat.columns[j], correlat.iloc[i, j]))

# Convert to DataFrame for easy handling
correlation_df = pd.DataFrame(correlation_pairs, columns=['Feature 1', 'Feature 2', 'Correlation'])

# Plotting the correlations as a bar chart
plt.figure(figsize=(10, 6))
bar_positions = np.arange(len(correlation_df))
plt.bar(bar_positions, correlation_df['Correlation'], color='skyblue')

# Set x-tick labels to show feature pairs
plt.xticks(bar_positions, correlation_df.apply(lambda x: f"{x['Feature 1']} & {x['Feature 2']}", axis=1), rotation=45, ha="right")
plt.ylabel('Correlation')
plt.title("Correlation between 'carat', 'depth', 'table', and 'price'")
plt.tight_layout()
plt.show()
```
